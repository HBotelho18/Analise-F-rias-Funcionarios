Aqui vamos analisar um dataset de 20 funcionarios de uma empresa

objetivo:


*   Analisar quais funcionarios estão com **férias em dia** e **férias vencidas**
*   Analisar o **impacto** financeiro que os funcionarios com férias vencidas pode gerar para a empresa
*   Analisar quais setores tem mais funcionarios com **ferias vencidas**

```
import pandas as pd

ferias = pd.read_csv('/content/drive/MyDrive/curriculos/dataset/dataset_funcionarios_veman.csv')
ferias.head(5)
```

**Ferias em dia**

```
ferias[ferias['anos_ferias_vencidas'] == 0].groupby(['nome','setor'])['anos_ferias_vencidas'].sum()
```

**Ferias vencidas**

```
ferias[ferias['anos_ferias_vencidas'] > 0].groupby(['nome','cargo'])['anos_ferias_vencidas'].sum().sort_values(ascending=False)
```

vamos calcular as ferias de cada funcionario e descobrir o impacto que cada funcionario com ferias vencidas gera a empresa



```
ferias['ds'] = ferias['salario'] * 2
ferias['ut'] = ferias['ds'] / 3
ferias['vf'] =  ferias['anos_ferias_vencidas'] * ferias['ds'] + ferias['ut']
ferias[ferias['anos_ferias_vencidas'] > 0].groupby(['nome','cargo','anos_ferias_vencidas'])['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```

```
ferias ['vencidas custo'] = ferias['ds'] = ferias['salario'] * 2
ferias['ut'] = ferias['ds'] / 3
ferias['vf'] =  ferias['anos_ferias_vencidas'] * ferias['ds'] + ferias['ut']
ferias[ferias['anos_ferias_vencidas'] > 0].groupby(['nome','cargo','anos_ferias_vencidas'])['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```

Totalizando

```
f"{ferias[ferias['anos_ferias_vencidas'] > 0]['vf'].sum():,.2f}"
```

```
ferias ['total custo'] = f"{ferias[ferias['anos_ferias_vencidas'] > 0]['vf'].sum():,.2f}"
```

Setor com maior custo em Férias

```
ferias.groupby('setor')['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```

```
ferias ['setor maior custo'] = ferias.groupby('setor')['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```

```
ferias.to_csv('dados_ferias.csv',index = False, sep = ';')
```
