Aqui vamos analisar um dataset de 20 funcionarios de uma empresa

objetivo:


*   Analisar quais funcionarios estão com **férias em dia** e **férias vencidas**
*   Analisar o **impacto** financeiro que os funcionarios com férias vencidas pode gerar para a empresa
*   Analisar quais setores tem mais funcionarios com **ferias vencidas**






```
import pandas as pd

url = '/content/dataset_funcionarios_veman.csv'
df = pd.read_csv(url)
df.head(21)
```

**Ferias em dia**

```
df[df['anos_ferias_vencidas'] == 0].groupby(['nome','setor'])['anos_ferias_vencidas'].sum()
```

**Ferias vencidas**

```
df[df['anos_ferias_vencidas'] > 0].groupby(['nome','cargo'])['anos_ferias_vencidas'].sum().sort_values(ascending=False)
```

vamos calcular as ferias de cada funcionario e descobrir o impacto que cada funcionario com ferias vencidas gera a empresa



```
df['ds'] = df['salario'] * 2
df['ut'] = df['ds'] / 3
df['vf'] =  df['anos_ferias_vencidas'] * df['ds'] + df['ut']
df[df['anos_ferias_vencidas'] > 0].groupby(['nome','cargo','anos_ferias_vencidas'])['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```

Totalizando

```
f"{df[df['anos_ferias_vencidas'] > 0]['vf'].sum():,.2f}"
```

Setor com maior custo em Férias

```
df.groupby('setor')['vf'].sum().sort_values(ascending=False).map('{:,.2f}'.format)
```
